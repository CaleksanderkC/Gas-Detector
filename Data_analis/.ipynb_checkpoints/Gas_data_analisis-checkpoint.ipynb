```python
import csv
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
from scipy import signal, stats
from scipy.optimize import curve_fit
from pandas_ods_reader import read_ods
```

```python
def read_lines(arr, start=4, end=7):
    temp = []

    for y, i in enumerate(arr):
        try:
            temp.append(int(i[start:end]))
        except:
            temp.append(temp[y-1])
    return np.array(temp)

def time2int(time):
    # Time as string
    h = int(time[:2])
    m = int(time[3:5])
    s = int(time[6:8])

    return  s + 60*m + 60*60*h
    
```

```python
R_L_ch4 = 5.01 #K OMEGA
R_L_co = 5.06 #K OMEGA


def u2rs_ch4(A_ch4):
    return R_L_ch4 * (A_ch4/(1023 - A_ch4))
# Rs_ch4 = R_L_ch4 * (A_ch4/(1023 - A_ch4))
# Rs_co = R_L_co * (A_co/(1023 - A_co))
```

<h2>ABB</h2>

```python
with open('Calibration_13_02_2025.csv', mode ='r') as file:
    csvFile = csv.reader(file)
    lines = np.array([i for i in csvFile])

DELTA = 196

Time_id = np.array([time2int(i) for i in [i[0][11:19] for i in lines[1:] ] ] ) - DELTA
CH4_ppm_t = np.array([float(i[6]) for i in lines[1:]])
CO_ppm_ = np.array([float(i[10]) for i in lines[1:]])


Time_real = np.arange(Time_id.min(), Time_id.max())
id_ = Time_id - Time_id.min()
CH4_ppm_ = np.zeros(len(Time_real)+1)
CH4_ppm_[:] = np.nan


for big_id, small_id in enumerate(id_):
    CH4_ppm_[small_id] = CH4_ppm_t[big_id]

del(lines)
del(id_)
```

```python
Time_del = Time_real.min()

def cal_time(time):
    return (time - Time_del)/60
```

```python
fig, ax1 = plt.subplots()

fig.set_size_inches(12,6)

clr = 'tab:red'
ax1.plot(cal_time(Time_real+DELTA), CH4_ppm_[:-1], label='Dane ABB', color=clr)
ax1.set_xlabel('Czas t (min)')
ax1.set_ylabel('PPM', color=clr)
ax1.tick_params(axis='y', labelcolor=clr)
ax1.grid(None)
ax1.legend(loc=2)

clr = 'tab:blue'
ax2 = ax1.twinx()
ax2.plot(cal_time(Time_2), u2rs_ch4(CH4_ppm_2[:-1]), label='Figaro TGS 2611', c=clr)
ax2.plot(cal_time(Time_1), u2rs_ch4(CH4_ppm_1), c=clr)
ax2.set_xlabel('Czas t (min)')
ax2.set_ylabel('R [kΩ]', color=clr)
ax2.tick_params(axis='y', labelcolor=clr)
ax2.legend(loc=0)

ax2.set_title('Porównanie pomiarów PPM CH4 z rezystancją czujnika Figaro bez uwzględnienia błędu czasowego na urządzeniu ABB.')
fig.tight_layout()
# plt.savefig('fig1.png')
plt.show()
```
