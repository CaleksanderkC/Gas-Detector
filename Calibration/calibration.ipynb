```python
import csv
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
from scipy import signal, stats
from scipy.optimize import curve_fit
from pandas_ods_reader import read_ods
```

```python
with open('Calib_data.csv', mode ='r')as file:
    csvFile = csv.reader(file)
    lines = [i for i in csvFile]
```

```python
Time = [i[1] for i in lines[1:]]
CH4_ppm = [float(i[2]) for i in lines[1:]]
CH4_d_ppm = [float(i[3]) for i in lines[1:]]
```

```python
start = 0
end = -1
for i, time in enumerate(Time):

    if time[12:17] == '14:16':
        start = i
    if time[12:17] == '14:50':
        end = i
```

```python
with open('PuTTY_out_CH4.txt', 'r') as f:
    k = [i for i in f]

A_1 = []

for i in k[::2]:
    A_1.append(int(i[4:]))

A_1 = np.array(A_1)
```

```python
x = np.arange(len(Time))
```

```python
Time_ = np.array(Time[start:end])

offset = 525
CH4_ppm_ = np.array(CH4_ppm[start:end])
CH4_ppm_ = CH4_ppm_[offset:offset+len(A_1)]

# CH4_d_ppm_ = np.array(CH4_d_ppm[start:end])
```

```python
s_ = 1200
e_ = 1250


plt.scatter(A_1[s_:e_], CH4_ppm_[s_:e_])
plt.yscale('log')
```

```python
s_3, e_3 = 708, 1100
s_2, e_2 = 1125, 1250
s_1, e_1 = 1300, len(A_1) - 100

def func_lin(x, a, b):
    return a*x+b

popt_3, pcov = curve_fit(func_lin, A_1[s_3:e_3],  np.log10(CH4_ppm_[s_3:e_3]))
popt_2, pcov = curve_fit(func_lin, A_1[s_2:e_2],  np.log10(CH4_ppm_[s_2:e_2]))
popt_1, pcov = curve_fit(func_lin, A_1[s_1:e_1],  np.log10(CH4_ppm_[s_1:e_1]))
```

```python
x_23 = -(popt_3[1] - popt_2[1])/(popt_3[0] - popt_2[0])
y_23 = popt_3[0] * x_23 + popt_3[1]
x_23, 10**y_23
```

```python
x_12 = -(popt_2[1] - popt_1[1])/(popt_2[0] - popt_1[0])
y_12 = popt_2[0] * x_12 + popt_2[1]
x_12/1024 * 5, x_23/1024 * 5
```

```python
def calibrate(arr):
    c = []
    for x in arr:
        if( x >= x_23):
            c.append( 10**func_lin(x, *popt_3))
        elif x >= x_12:
            c.append( 10**func_lin(x, *popt_2))
        else :
            c.append( 10**func_lin(x, *popt_1))
    return np.array(c)
    
```
